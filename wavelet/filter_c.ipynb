```python
signal = Signal.from_path("audio.wav")
signal.displayAudio()
```

```python
z = sympy.symbols('z')
```

$
H_0(z) = \frac{1}{4}(1 + z^{-1})^2
$

```python
H0_poly = 1/4 *(1 + z**(-1))**2
```

$
F_0(z) = -\frac{1}{4}(1+z^{-1})^2(2+\sqrt{3}-z^{-1})(2-\sqrt{3}-z^{-1})
$

```python
F0_poly = -1 / 4 * (2 - sympy.sqrt(3) - z ** (-1)) * (2 + sympy.sqrt(3) - z ** (-1)) * (1 + z ** (-1))**2
```

```python
H0, F0, H1, F1 = Signal.from_polynom(H0_poly, F0_poly, z)
```

$
y_0[n]=(z\ast H_0)[2n]
$

$
y_1[n]=(z\ast H_1)[2n]
$

```python
y0, y1 = signal.Analysis(H0, H1)
```
